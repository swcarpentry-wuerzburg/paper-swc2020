1D plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = numpy.loadtxt(fname='data/inflammation-01.csv', delimiter = ',')
```

```python
data_mean = numpy.mean(data,axis = 0)
```

```python
plt.plot(data_mean)
plt.show()
```

```python
max_plot_log = plt.plot(data_mean)
plt.yscale("log")
plt.show()
```

```python
data_std = numpy.std(data, axis = 0)
fig = plt.figure(figsize=(20,10))
plt.errorbar(x=np.linspace(0,40,40), y=data_mean, yerr=data_std, xerr=np.zeros(40), capsize=5, marker="o")
plt.show()
```

```python
data_std = numpy.std(data, axis = 0)
fig = plt.figure(figsize=(20,10))
ax = fig.gca()
ax.errorbar(x=np.linspace(0,40,40), y=data_mean, yerr=data_std, xerr=np.zeros(40), capsize=5, marker="o")
plt.yscale("log")
ax.set_ylim([0.25,18])
plt.show()
```

3D plots

```python
X = np.linspace(0, 40, num=40)
Y = np.linspace(0, 60, num=60)
X, Y = np.meshgrid(X,Y)
```

```python
fig = plt.figure(figsize=(15,15))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X,Y,data, cmap = "hot")
fig.colorbar(surf, shrink=0.5)
plt.show()
```

```python
import math
```

```python
coords = np.zeros((40,60,2))

rs = np.linspace(-1,1,num=40)
increment_angle = math.pi/60

for i in range(60):
    angle = i*increment_angle
    coords[:,i,:] =  np.resize(rs,(40,1))*np.resize(np.array([math.sin(angle), math.cos(angle)]), (40,2))
```

```python
numspokes = 60
fig = plt.figure(figsize=(10.0, 10.0))
ax = fig.gca()
ax.plot(coords[:,0:numspokes,0], coords[:,0:numspokes,1], marker = ".")
ax.set_ylim([-1.05,1.05])
ax.set_xlim([-1.05,1.05])
plt.show()
```

```python
numspokes = 60
fig = plt.figure(figsize=(10,10))
ax = fig.gca()
xcoords_1d = np.reshape(coords[:,0:numspokes,0], (40*numspokes))
ycoords_1d = np.reshape(coords[:,0:numspokes,1], (40*numspokes))
data_1d = np.reshape(np.transpose(data[0:numspokes,:]),(40*numspokes))


ax.scatter(xcoords_1d, ycoords_1d, c = data_1d, cmap="hot", marker="o", s=100)
plt.plot();
```
